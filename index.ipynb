## Step 1: Read the dataset and inspect its columns and 5-point statistics

```python
# Load necessary libraries and import the data
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
plt.style.use('ggplot')

df = pd.read_csv('advertising.csv', index_col=0)
```

```python
# Check the columns and first few rows
df.head()
```

```python
# Get the 5-point statistics for data 
df.describe()
```

```python
# This dataset contains 4 features: TV, Radio, Newspaper, and Sales.
# TIt is assumed that these are monetary values, the budget spent on each
# advertising stream. The target variable here wouldl be sales, as it is 
# dependent on the amount spent per medium.  
# There seems to be some outliers on the low-end of the TV feature, as well
# as on the high-end of the newspaper feature.
# The most money spent seems to be in TV, with the least being radio.
```

## Step 2: Plot histograms with kde overlay to check the distribution of the predictors

```python
# For all the variables, check distribution by creating a histogram with kde
for column in df:
    df[column].plot.hist(normed=True, label= column+ ' histogram');
    df[column].plot.kde(label = column+ ' KDE');
    plt.legend();
    plt.show();
```

```python
# The data for TV, radio, and sales seem to follow a normal distribution curve.
# The newspaper data seems to be positively skewed though.
```


## Step 3: Test for the linearity assumption 

Use scatterplots to plot each predictor against the target variable

```python
# visualize the relationship between the preditors and the target using scatterplots
fig, axes = plt.subplots(1,3, sharey=True, figsize=(18,6))

for idx, channel in enumerate(['TV', 'radio', 'newspaper']):
    df.plot(kind='scatter', x=channel, y='sales', ax=axes[idx], label=channel)

plt.legend();
plt.show();
```

```python
# TV seems to be the most linear compared to sales of the three features.
# There is a bit of linearity with radio and sales but it varies quite a bit.
# There does not seem to be any linearity between newspaper and sales though.
# OLS assumptions hold for TV and we will try radio. However no go for newspaper!
```

### Conclusion so far

Based on above initial checks, we can confidently say that TV and radio appear to be good predictors for our regression analysis. Newspaper is very heavily skewed and also doesnt show any clear linear relationship with the target. 
> We'll move ahead with our analysis using TV and radio, and rule out newspaper because we believe it violates OLS assumptions

Note: Kurtosis can be dealt with using techniques like log normalization to "push" the peak towards the center of distribution. You'll learn about this later on.

## Step 4: Run a simple regression in Statsmodels with TV as a predictor

```python
# import libraries
# build the formula 
f = 'sales ~ TV'
# create a fitted model in one line
model = ols(formula=f, data=df).fit()
```

## Step 5: Get Regression Diagnostics Summary

```python
model.summary()
```

Note here that the coefficients represent associations, not causations

## Step 6:  Draw a prediction line with data points on a scatter plot for X (TV) and Y (Sales)

Hint: You can use the `model.predict()` function to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
TV_min_max_df = pd.DataFrame({'TV': [df.TV.min(), df.TV.max()]})
print(TV_min_max_df.head())

# make predictions for those x values and store them
preds = model.predict(TV_min_max_df)
print (preds)

# first, plot the observed data and the least squares line
df.plot(kind='scatter', x='TV', y='sales');
plt.plot(TV_min_max_df, preds, c='green', linewidth=2);
plt.show();
```

## Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(15,8));
fig = sm.graphics.plot_regress_exog(model, 'TV', fig=fig);
plt.show();
```

```python
# There seems to be a lot of heteroscedacity in these plots due to the cone
# shape that we see in the first two plots (only relevent plots due to there 
# only being one independent variable)
```

## Step 8: Check the normality assumptions by creating a QQ-plot

```python
# Code for QQ-plot here
import scipy.stats as stats

residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
fig.show();
```

```python
# All of the residual data points shown in the above QQ-plot are very close to
# the line, indicating that the residuals do follow a normal distribution.
```

## Step 9: Repeat the above for radio and record your observations

```python
f = 'sales ~ radio'
model = ols(formula=f, data=df).fit()
```

```python
model.summary()
```

```python
# create a DataFrame with the minimum and maximum values of radio
radio_min_max_df = pd.DataFrame({'radio': [df.radio.min(), df.radio.max()]})
print(radio_min_max_df.head())

# make predictions for those x values and store them
preds = model.predict(radio_min_max_df)
print (preds)

# first, plot the observed data and the least squares line
df.plot(kind='scatter', x='radio', y='sales');
plt.plot(radio_min_max_df, preds, c='green', linewidth=2);
plt.show();
```

```python
fig = plt.figure(figsize=(15,8));
fig = sm.graphics.plot_regress_exog(model, 'radio', fig=fig);
plt.show();
```

```python
# Code for QQ-plot here
import scipy.stats as stats

residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
fig.show();
```

```python
# Each of the visualizations above for sales vs. radio perform slightly worse
# than sales vs. TV.  The R-squared value is half at 0.3, there is even more
# heteroscedacity than TV, and the residuals do not follow as normal of a 
# distribution as the TV residuals do.

# Therefore, I would conclude that between the two of them, TV seems to be a 
# better predictor of sales than radio!
```
